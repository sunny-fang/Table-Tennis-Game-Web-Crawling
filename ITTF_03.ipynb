```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
from selenium.webdriver.support.ui import WebDriverWait
from tqdm.auto import tqdm
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep
```

```python
def get_player_matches(player,page_index):
    # 開啟webdriver
    driver = webdriver.Chrome()
    driver.implicitly_wait(30) # 隱性等待，最長30秒
    driver.get('https://results.ittf.link/index.php?option=com_fabrik&view=list&listid=99&Itemid=158')
    driver.maximize_window()
    # 處理下拉式選單
    select = Select(driver.find_element_by_name('fabrik___filter[list_99_com_fabrik_99][value][0]'))
    select.select_by_visible_text(u"CHINA")
    select = Select(driver.find_element_by_name('fabrik___filter[list_99_com_fabrik_99][value][1]'))
    select.select_by_visible_text(u"Male")
    # 處理按鈕"GO"
    #WebDriverWait(driver, 20).until(lambda driver:driver.find_element_by_name("filter"))
    driver.find_element_by_name("filter").click()
    # 處理按鈕:指定點選"page_index"這個數字的按鈕
    wait = WebDriverWait(driver, 20)
    element = wait.until(EC.element_to_be_clickable((By.LINK_TEXT, page_index)))
    element.click()
    # 處理按鈕:指定點選"FAN Zhendong"這個字的按鈕
    #WebDriverWait(driver, 20).until(lambda driver:driver.find_element_by_link_text(player))
    driver.find_element_by_link_text(player).click()
    #WebDriverWait(driver, 20).until(lambda driver:driver.find_element_by_id('limit47'))
    sleep(5)
    select = Select(driver.find_element_by_id('limit47'))
    select.select_by_visible_text(u"all")
    tournament = [];player_A = [];player_B = [];player_X = [];player_Y = [];event = [];round_a = [];res_A = [];res_X = [];winner = [];winner_dbl = []
    testsoup = BeautifulSoup(driver.page_source, "html.parser")
    aa = testsoup.findAll('tr', {'id':re.compile('^(list_47_com_fabrik_47_row_)')})
    for index, row in enumerate(aa):
        tournament.append(aa[index].find_next('td',{'class':'vw_tournaments___tournament fabrik_element fabrik_list_47_group_71'}).text.replace('\n','').replace('\t',''))
        player_A.append(aa[index].find_next('td',{'class':'fab_matches___name_a fabrik_element fabrik_list_47_group_50'}).text.replace('\n','').replace('\t',''))
        player_B.append(aa[index].find_next('td',{'class':'fab_matches___name_b fabrik_element fabrik_list_47_group_50'}).text.replace('\n','').replace('\t',''))
        player_X.append(aa[index].find_next('td',{'class':'fab_matches___name_x fabrik_element fabrik_list_47_group_50'}).text.replace('\n','').replace('\t',''))
        player_Y.append(aa[index].find_next('td',{'class':'fab_matches___name_y fabrik_element fabrik_list_47_group_50'}).text.replace('\n','').replace('\t',''))
        event.append(aa[index].find_next('td',{'class':'fab_matches___event fabrik_element fabrik_list_47_group_50'}).text.replace('\n','').replace('\t',''))
        round_a.append(aa[index].find_next('td',{'class':'fab_matches___round fabrik_element fabrik_list_47_group_50 integer'}).text.replace('\n','').replace('\t',''))
        res_A.append(aa[index].find_next('td',{'class':'fab_matches___res_a fabrik_element fabrik_list_47_group_50 integer'}).text.replace('\n','').replace('\t',''))
        res_X.append(aa[index].find_next('td',{'class':'fab_matches___res_x fabrik_element fabrik_list_47_group_50 integer'}).text.replace('\n','').replace('\t',''))
        winner.append(aa[index].find_next('td',{'class':'fab_matches___winner_name fabrik_element fabrik_list_47_group_50'}).text.replace('\n','').replace('\t',''))
        winner_dbl.append(aa[index].find_next('td',{'class':'fab_matches___winners_names fabrik_element fabrik_list_47_group_50'}).text.replace('\n','').replace('\t',''))

    driver.close()
    df = pd.DataFrame({'tournament':tournament, 'player_A':player_A, 'player_B':player_B, 'player_X':player_X, 'player_Y':player_Y, 
                      'event':event, 'round':round_a, 'res_A':res_A ,'res_X':res_X, 'winner':winner, 'winner_dbl':winner_dbl})
    df.to_csv(player+'.csv', index=False)

    return df
```
