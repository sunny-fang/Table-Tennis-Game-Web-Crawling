```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
from selenium.webdriver.support.ui import WebDriverWait
from tqdm.auto import tqdm
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep
```

```python
driver = webdriver.Chrome()
driver.get('https://results.ittf.link/index.php?option=com_fabrik&view=list&listid=119&Itemid=293&resetfilters=0&clearordering=0&clearfilters=0')
```

```python
select = Select(driver.find_element_by_name('fabrik___filter[list_119_com_fabrik_119][value][2]'))
select.select_by_visible_text(u"Male")
driver.find_element_by_name("filter").click()
```

```python
i=5
driver.find_element_by_link_text(str(i)).click()
```

```python
name=[]
for i in range(1,15):
    driver.find_element_by_link_text(str(i)).click()
    WebDriverWait(driver, 20).until(lambda driver:driver.find_element_by_link_text(str(i)))
    testsoup = BeautifulSoup(driver.page_source, "html.parser")
    for tour in testsoup.findAll('a',{'class':'fabrik___rowlink'}):
        name.append(tour.text.replace('\n','').replace('\t',''))
    sleep(2)
```
