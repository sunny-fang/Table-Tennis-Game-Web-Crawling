```python
import urllib
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import requests
```

```python
def get_web_page(url):
    resp = requests.get(url)
    resp.encoding = 'utf-8'
    
    if resp.status_code != 200:# 是200就代表成功拜訪網站
        print('Invalid url:', resp.url)
        return None
    else:
        return resp.text
url = 'https://results.ittf.link/index.php?option=com_fabrik&view=list&listid=47&Itemid=158&resetfilters=0&clearordering=0&clearfilters=0'
resp = get_web_page(url)
soup = BeautifulSoup(resp, 'html.parser')
```
